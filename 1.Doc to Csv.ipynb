```python
import os
import fitz  #  for PDFs
import docx2txt  # For DOCX files
import win32com.client  # For DOC files
import pandas as pd
import numpy as np
```

## Exctract Text from Dataset

```python
file_path = [] 
category = []
directory = r'C:\Users\sanke\OneDrive\Desktop\Excelr\Projects\DS Project\Project 3 Resume classification\Resumes_Docx\React Developer'

for i in os.listdir(directory):
    full_path = os.path.join(directory, i)
    
    if i.endswith('.docx'):
        # Handle DOCX files
        text = docx2txt.process(full_path)
    
    elif i.endswith('.pdf'):
        # Handle PDF files
        pdf_doc = fitz.open(full_path)
        text = ""
        for page in pdf_doc:
            text += page.get_text()
    
    elif i.endswith('.doc'):
        # Handle DOC files using win32com (Windows only)
        word = win32com.client.Dispatch("Word.Application")
        doc = word.Documents.Open(full_path)
        text = doc.Content.Text
        doc.Close()
    
    file_path.append(text)
    category.append('React Developer Resume')
```
